**{:.4f} --> Serve para definir o print com 4 casas decimais **

```python
Lista_notas=[3.4, 4.5, 7.0, 8.0, 5.5, 7.5]
soma=0
for notas in Lista_notas:
  soma= soma + notas
  media=soma/len(Lista_notas)
print('Média=', '{:.4f}'.format(media))
```
