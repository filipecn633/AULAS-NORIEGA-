Len mostra a quantia da lista. nesse caso, foi usado para calcular a média

```python
Lista=[4.5,5.6,7.7,3.5,8.9,9.0,4.5]
soma=0
for nota in Lista:
  soma=soma+nota
media=soma/len(Lista)
print("A média é: ",'{:.4f}'.format(media))
```

Aqui abaixo foi feita o mesmo calculo de media mas com funções

```python
#criação da função para calculo de media de uma lista
def CalcMedia(lista):
  soma=0
  for nota in lista:
    soma=soma+nota
  return soma/len(lista)
```

```python
Lista1=[4.5,5.6,7.7,3.5,8.9,9.0,4.5]

media=CalcMedia(Lista1)
print("A média é: ",'{:.4f}'.format(media))

```
